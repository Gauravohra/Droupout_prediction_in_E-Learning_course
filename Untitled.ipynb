```python
import pandas as pd
import random

# Create an empty DataFrame
data = pd.DataFrame(columns=[
    'student_id', 'age', 'gender', 'course_type', 'login_count', 'discussion_posts',
    'num_assignments', 'total_assignment_score', 'num_homeworks', 'num_homework_submissions',
    'dropout', 'performance'
])

# Generate 200+ data points
data_list = []  # Create a list to hold data dictionaries

for student_id in range(1, 201):
    age = random.randint(18, 30)
    gender = random.choice(['Male', 'Female'])
    course_type = random.choice(['Math', 'History', 'Science'])
    login_count = random.randint(5, 15)
    discussion_posts = random.randint(0, 5)
    num_assignments = random.randint(3, 6)
    total_assignment_score = random.randint(200, 500)
    num_homeworks = random.randint(4, 6)
    num_homework_submissions = random.randint(3, num_homeworks)
    dropout = random.randint(0, 1)
    performance = random.randint(60, 90)

    data_dict = {
        'student_id': student_id,
        'age': age,
        'gender': gender,
        'course_type': course_type,
        'login_count': login_count,
        'discussion_posts': discussion_posts,
        'num_assignments': num_assignments,
        'total_assignment_score': total_assignment_score,
        'num_homeworks': num_homeworks,
        'num_homework_submissions': num_homework_submissions,
        'dropout': dropout,
        'performance': performance
    }

    data_list.append(data_dict)

# Concatenate the list of data dictionaries into a DataFrame
data = pd.concat([data, pd.DataFrame(data_list)])

# Save the DataFrame to a CSV file
data.to_csv('elearning_data_large.csv', index=False)

```

```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv("elearning_data_large.csv")
```

```python
# Drop rows with missing values
data.dropna(inplace=True)

# Encode categorical variables (if any)
data = pd.get_dummies(data, columns=["gender", "course_type"])
```

```python
# Create new features for dropout prediction
data["logins_in_first_week"] = data["login_count"] <= 7
data["discussion_participation"] = data["discussion_posts"] > 0
data["average_assignment_score"] = data["total_assignment_score"] / data["num_assignments"]

# Create new features for performance prediction
data["homework_completion_rate"] = data["num_homework_submissions"] / data["num_homeworks"]
```

```python
from sklearn.model_selection import train_test_split

X_dropout = data[["logins_in_first_week", "discussion_participation", "average_assignment_score"]]
y_dropout = data["dropout"]

X_perf = data[["logins_in_first_week", "discussion_participation", "homework_completion_rate"]]
y_perf = data["performance"]

# Split the data for both tasks
X_train_dropout, X_test_dropout, y_train_dropout, y_test_dropout = train_test_split(X_dropout, y_dropout, test_size=0.3, random_state=42)
X_train_perf, X_test_perf, y_train_perf, y_test_perf = train_test_split(X_perf, y_perf, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model_dropout = LogisticRegression(random_state=42)
```

```python
model_dropout.fit(X_train_dropout, y_train_dropout)

# Make predictions
y_pred_dropout = model_dropout.predict(X_test_dropout)

# Evaluate the model for dropout prediction
from sklearn.metrics import accuracy_score, classification_report

accuracy_dropout = accuracy_score(y_test_dropout, y_pred_dropout)
report_dropout = classification_report(y_test_dropout, y_pred_dropout)

print("Dropout Prediction:")
print(f'Accuracy: {accuracy_dropout:.2f}')
print(report_dropout)
```
